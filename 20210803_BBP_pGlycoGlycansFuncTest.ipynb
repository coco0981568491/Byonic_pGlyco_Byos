```python
## THIS IS FOR BYONIC (USE SULFATION AS TESTING DATA)
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
from matplotlib import ticker
from matplotlib.ticker import ScalarFormatter
from IPython.display import display, HTML
import re # finding specific patterns in str
import textwrap # split text into equal parts
import collections # return repeated items in list
from collections import OrderedDict
import time
from datetime import datetime # attach current date to export filename
import sys
import ast # convert str back to tuple/list/int, etc
import matplotlib.cm as cm
import matplotlib.colors as mcolors
import string

def ReadExcelandFixCols(filename):
    df = pd.read_excel('%s.xlsx'%filename, header = 0)
    df = df.fillna('N/A')
    fixed_colname = [i.replace('_x000D_', '\r') if '_x000D_' in i else i for i in df.columns]
    df.columns = fixed_colname
    return df
def move_df(df, move_col, insert_left_col):
    move_df = df[move_col]
    df.drop(labels=[move_col], axis=1, inplace = True)
    df.insert(df.columns.get_loc(insert_left_col) + 1, move_col, move_df)
def get_key(val):
    for key, value in mods_NameToNum.items():
        if val == value:
            return key
    
# pass in the modification numbers in list (all the reuslts from byonic will be divided by the nums in this lst to find the most probable modification)
# the reuslts in byonic are calculated from the passed-in nums, so should be completely divided if it's the correct mods
mods_name = ['SO3', 'Phospho', 'KDN'] # PLEASE PASS IN W/ THE SAME ORDER AS MODS_NUM (CAN ONLY PASS IN THE FOLLOWING CHOICES: 'SO3', 'Phospho', 'KDN')
mods_num = [79.9568123, 79.966331, 250.0688572] # PLEASE PASS IN W/ THE SAME ORDER AS MODS_NAME

# construct relation btw mods_name & mods_num
mods_NameToNum = dict(zip(mods_name, mods_num))


df = ReadExcelandFixCols('sulfation_test')

df['GlycansModNum'] = df['Glycans'].str.replace('; ', ' ')
df['GlycansModNum'] = ['N/A' if i==[] else np.array([float(j) for j in i]) if len(i)>1 else float(i[0]) for i in df['GlycansModNum'].str.findall(r'\d+\.\d+')]


# start finding the possible mods
for num in mods_num:
    df['/%s'%(get_key(num))] = df['GlycansModNum'].apply(lambda x: list(x/num) if type(x)==np.ndarray else x/num if type(x)==float else None)
    df['/%s'%(get_key(num))] = df['/%s'%(get_key(num))].apply(lambda x: '' if x is not None and type(x)==float and x.is_integer()==False else '%s(%s)'%(get_key(num), int(x)) if x is not None and type(x)==float and x.is_integer()==True else np.char.array(['' if j.is_integer()==False else '%s(%s)'%(get_key(num), int(j)) for j in x]) if x is not None and type(x)==list else '')
df['final'] = df[['/%s'%(name) for name in mods_name]].sum(axis=1)

df['PureGlycans'] = df['Glycans'].str.replace(r' \d+\.\d+', '', regex=True).to_frame()
df['PureGlycans'] = [i[0] if len(i)==1 else np.char.array(i) for i in df['PureGlycans'].str.split(';')]

df['FinalGlycans'] = df['PureGlycans'] + df['final']
df['FinalGlycans'] = [';'.join(list(i)) if type(i)==np.chararray else i for i in df['FinalGlycans']]

# numpy.chararray
# df.to_excel('GlycansModNum_test2.xlsx', index=False)
```
